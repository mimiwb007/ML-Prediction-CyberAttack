```python
#read the survey dataset
#might take some time as the data is large in size
import pandas as pd

df = pd.read_excel(r"Data.xlsx")
df.head()
```

```python
#copy the data to another dataframe
d2=df.copy()
d2.head()
```

```python
#do some transformations for the categorical columns for the prediction model
l1=['12-17 years old','Under 12 years old']
l2 =['18-24 years old','25-34 years old','35-44 years old']
l3 = ['45-54 years old','55-64 years old','65-74 years old']
l4=['75 years or older']
l5=["Other - Don't Know - Prefer Not To Answer - N/A"]
#age group
def mak(col):
    if col in l1:
         return 'Under 17 years'
    elif col in l2:
          return '18-45 years old'
    elif col in l3:
          return '45-75 years old'
    elif col in l4:
         return '75 years or older'
    elif col in l5:
        return 'NA'
d2['Age Group'] = d2['Age Group'].apply(mak)      
```

```python
#do some transformations for the categorical columns for the prediction model

def mmakk(col):
    if col!="Other - Don't Know - Prefer Not To Answer - N/A":
        p = str(col).replace('€','')
        p1= p.split('-')
        h = (float(p1[0]) + float(p1[1]))/2
        return h
    else:
        return col

d2['Household Income'] = d2['Household Income'].apply(mmakk)
```

```python
#do some transformations for the categorical columns for the prediction model
d2.loc[d2['Household Income']=="Other - Don't Know - Prefer Not To Answer - N/A",'Household Income']=49.99833397268536
```

```python
#do some transformations for the categorical columns for the prediction model
e1=['Unemployed and not currently looking for work','Homemaker','Retired','Unemployed and currently looking for work','Student','Retired','Unable to work']
e2 =['Employed full time (40 or more hours per week)','Employed part time (up to 39 hours per week)']
e3 = ['Self-employed']
def emp(col):
    if col in e1:
       return 'Unemployed'
    elif col in e2:
        return 'Employed in Office'
    elif col in e3:
        return 'Self-Employed'
    elif col=="Other - Don't Know - Prefer Not To Answer - N/A":
        return col
    
    
d2['Employment Status']=d2['Employment Status'].apply(emp)
    
```

```python
#do some transformations for the categorical columns for the prediction model

enc ={'High school degree or equivalent (e.g. GED)':1,'Less than a high school diploma':0,'Some college, no degree':2,'Professional degree (e.g. MD, DDS, DVM)':3,'Bachelor’s degree (e.g. BA, BS)':4,'Associate degree (e.g. AA, AS)':5,'Master’s degree (e.g. MA, MS, MEd)':6,'Doctorate (e.g. PhD, EdD)':7}
d2['Level of Education']=d2['Level of Education'].map(enc)
#do some transformations for the categorical columns for the prediction model
d2['Level of Education']=d2['Level of Education'].fillna(d2['Level of Education'].mode().iloc[0])
```

```python
#do some transformations for the categorical columns for the prediction model

ll= ['Under 100 people','over 100.000.000 people']
def rem(col):
    if col not in ll:
        p= col.replace('.','')
        p1 = p.replace('people','')
        p2 = p1.split('-')
        kk = (int(p2[0])+int(p2[1]))/2
        return kk
    elif col == 'Under 100 people':
        return 100
    elif col=='over 100.000.000 people':
        return 100000000

d2['Place of Residence Population']=d2['Place of Residence Population'].apply(rem)
```

```python
#select the columns for model
#not included the country column here as it has too many categories and might remove the accuracy of the model
cat_col = ['Gender', 'Age Group', 'Marital Status', 'Employment Status',
       'Household Income', 'Level of Education',
       'Place of Residence Population',
       'Place of the last Incident', 'Day', 'Month', 'Type of Malware',
       'Economic Impacts', 'Technical Impacts', 'Type of Device',
       'Type of Operating System', 'Malware Protection','Αttacked by a Cyber Attack']
d2=d2[cat_col]
```

```python
#make one hot encoding for the X variables that are categorical
train = d2.drop('Αttacked by a Cyber Attack',axis=1)
d3 = pd.get_dummies(train)
#define the X and Y
X = d3.copy()
y = d2['Αttacked by a Cyber Attack']
```

```python
#do a train test split for checking the model's validation accuracy
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
X_t, X_tt, y_t, y_tt = train_test_split(X, y, test_size=.25, random_state=150303,stratify=y,shuffle=True)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(random_state=43).fit(X, y)
log.fit(X, y)
```

```python
#AdaBoost classifier
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier(n_estimators=100, random_state=0)
ada.fit(X, y)
```

```python
#Random Forest classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=2, random_state=0)
rf.fit(X, y)
```

```python
#save adaboost predictions
pred=ada.predict(X)
pred=pd.DataFrame(pred)
pred['True']=d2['Αttacked by a Cyber Attack']
pred.columns=['y_pred','y_true']
k1 = k+'\\Ada.csv'
pred.to_csv(k1,index=False)
```
